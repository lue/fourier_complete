```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wn = np.random.normal(0,1,1024)

plt.figure(0)
plt.plot(wn)

ft_wn = np.fft.rfft(wn)
freq = np.fft.rfftfreq(1024)

plt.figure(1)
plt.plot(freq, ft_wn, '.')
```

```python
scale = np.abs(freq)**-1
scale[0] = 1
scale[np.abs(freq)>0.01] = 0
ft_signal = ft_wn * scale
signal = np.real(np.fft.irfft(ft_signal))

plt.figure(2)
plt.plot(signal)

signal_o = signal.copy()
signal_o[512-64:512+64] = 0

signal_p = signal.copy()
signal_p -= signal_o

plt.plot(signal_o)
plt.plot(signal_p)
```

```python
ft_signal_o = np.fft.rfft(signal_o)
ft_signal_p = np.fft.rfft(signal_p)
print(ft_signal_o[1])
print(ft_signal_p[1])
print(ft_signal[1])
```

```python
plt.figure(11)
circle1 = plt.Circle((0,0),np.abs(ft_signal_o[1]), fill=False)
plt.gca().add_artist(circle1)
plt.plot(np.real(ft_signal_o[1]), np.imag(ft_signal_o[1]) ,'o')
plt.plot(np.real(ft_signal[1]), np.imag(ft_signal[1]) ,'o')
circle2 = plt.Circle((np.real(ft_signal_o[1]), np.imag(ft_signal_o[1])),np.abs(ft_signal_p[1]), fill=False)
plt.gca().add_artist(circle2)

plt.axis('equal')
plt.draw()
```
