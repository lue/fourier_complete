```python
def triginterp(xi,x,y):
    # % TRIGINTERP Trigonometric interpolation.
    # % Input:
    # %   xi  evaluation points for the interpolant (vector)
    # %   x   equispaced interpolation nodes (vector, length N)
    # %   y   interpolation values (vector, length N)
    # % Output:
    # %   P   values of the trigonometric interpolant (vector)
    N = len(x)
    #% Adjust the spacing of the given independent variable.
    h = 2./N
    scale = (x[1]-x[0]) / h
    x = x/scale
    xi = xi/scale
    # % Evaluate interpolant.
    P = np.zeros(len(xi))
    for k in range(N): #1:N
        P = P + y[k]*trigcardinal(xi-x[k],N)
    return P

def trigcardinal(x,N):
    # ws = warning('off','MATLAB:divideByZero');
    # % Form is different for even and odd N.
    if np.mod(N,2)==1:#   % odd
        tau = np.sin(N*np.pi*x/2.) / (N*np.sin(np.pi*x/2.))
    else: #             % even
        tau = np.sin(N*np.pi*x/2.) / (N*np.tan(np.pi*x/2.))
    # end
    # warning(ws)
    tau[x==0] = 1# ;     % fix value at x=0
    return tau
```

```python
def dft_nan(x):
    """Compute the discrete Fourier Transform of the 1D array x"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    print(M)
#     return np.dot(M, x)
    X = np.zeros(len(x))
    for i in range(len(X)):
        X[i] = np.nansum(M[:,i]*x)
    return X
```

```python
def dft_fill(x):
    """Compute the discrete Fourier Transform of the 1D array x"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    Md = np.mod(k * n, N)
    M = np.exp(-2j * np.pi * Md / N)
    print(M)
#     return np.dot(M, x)
    X = np.zeros(len(x))
    for i in range(len(X)):
        print(k[i], len(np.unique(Md[:,i])))
        X[i] = np.nansum(M[:,i]*x)
    return X
```
