```python
import scipy as sp
from scipy.signal import fftconvolve, convolve
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

```

```python
N=2**10
wn = np.random.normal(0,1,size=[N])
```

```python

freq = np.fft.fftfreq(1024)

scale = np.abs(freq)**-1
scale[0] = 1
scale[np.abs(freq)>0.05] = 0
scale[np.abs(freq)<0.01] = 0

co = np.fft.ifft(freq)
# co = np.exp(-(np.arange(-256,256)/256)**1)
# co = np.ones(512)
```

```python
plt.figure(1)
plt.clf()
plt.plot(np.real(co))
```

```python
con = fftconvolve(wn, co, "same")
```

```python
plt.clf()
plt.plot(con)
```
